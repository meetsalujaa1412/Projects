### Starting with Exploration and Hypothesis Development

#### Business Problem Statement
Background

A retail brand has partnered with a marketing consulting firm to improve the effectiveness of its customer marketing strategy. Currently, the brand relies on broad, uniform marketing campaigns (e.g., generic discounts, mass promotions) applied across its entire customer base. While this approach ensures coverage, it results in inefficient marketing spend and inconsistent customer engagement.

The brand has access to basic customer demographic and behavioral data, including age, gender, annual income, and a composite spending score that reflects customer purchasing behavior. However, this data has not yet been leveraged to understand distinct customer behavior patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Drawing relationship between Age,Income and Spending Score

```python
plt.figure(figsize=(25,7))
plt.subplot(1,3,1)
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Scre (1-100)')

plt.subplot(1,3,2)
plt.scatter(df['Age'], df['Annual Income (k$)'])
plt.ylabel('Annual Income')
plt.xlabel('Age')

plt.subplot(1,3,3)
plt.scatter(df['Age'], df['Spending Score (1-100)'])
plt.ylabel('Spending Score (1-100)')
plt.xlabel('Age')
plt.show()
```

#### Preliminary Insights
1. The spending behaviour showcases the median spending score and income between 40-60 and 60-80k respectively and people with high income don't tend to always spend more and there are customers who do have high income but spend significantly less and vice versa.
2. The people in the age between 20-40 have a high spending score compared to people from 45-70 who have low to median spending score around 60.
3. Most of the customers have an annual income <= 80k 


#### Refined Insights

1. Customers with similar income levels exhibit very different spending behaviors. High income does not consistently translate into high spending, and some lower-income customers demonstrate high spending engagement.
2. Younger customers (approximately ages 2040) tend to show higher spending engagement compared to older customers (4570), who cluster around low to medium spending levels.
3. The majority of customers fall below an annual income of approximately 80k, suggesting that high-income customers represent a smaller but potentially important segment.

#### Conclusion from Exploratory Analysis

Customer behavior varies significantly across income and age groups, and spending patterns are not uniformly explained by any single variable. The presence of distinct behavioral patterns suggests that customer segmentation may help the marketing team better tailor strategies and allocate marketing resources efficiently.
